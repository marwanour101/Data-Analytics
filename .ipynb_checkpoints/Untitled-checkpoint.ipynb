```python
import pandas as pd
```

```python
df = pd.read_csv("bank-direct-marketing-campaigns.csv" , encoding='latin-1')
df.info()
```

```python
df.isna().sum()
# there is no nulls to remove
```

```python
df.fillna(0) 
# filling missing values by zero
```

```python
import scipy.stats as stats
Q1 = df.quantile(q = .25)
Q3 = df.quantile(q = .75)
IQR = df.apply(stats.iqr)
```

```python
df = df[~(( df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```
